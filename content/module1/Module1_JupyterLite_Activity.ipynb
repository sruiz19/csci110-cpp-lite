# Running Code

To run a code cell:

1. Click inside the cell
2. Press `Shift + Enter`

# Restarting the Kernel

If your code stops working or you see a "redefinition of main" error:

1. Go to `Kernel`
2. Select `Restart Kernel and Clear Outputs`
3. Run your code again

# Downloading Your Work

Your notebook is stored in your browser while using JupyterLite.

To save your notebook to your computer:

1. Go to `File`
2. Select `Download`
3. Save the `.ipynb` file
